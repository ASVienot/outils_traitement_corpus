```python
import pandas as pd 
import spacy

nlp = spacy.load('fr_core_news_sm')

data = pd.read_csv("../data/.csv")
data['type'] = data['title'].str[0]



def lemmatization(body):
    doc = nlp(body)
    lemma = [token.lemma_ for token in doc]
    return lemma

# Apply lemmatization to the 'body' column and create a new column 'lemma'
data['lemma'] = data['body'].apply(lemmatization)


data
```
